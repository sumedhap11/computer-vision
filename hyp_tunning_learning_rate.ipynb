```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-1,1,100)
y=2*x+1
```

```python
def train_model(lr,epochs=50):
    w, b = 0.0,0.0
    loss_values =[]
    
    for epoch in range(epochs):
        y_pred =w * x+b
        loss=np.mean((y-y_pred)**2)
        loss_values.append(loss)
        
        dw=np.mean(2*(y_pred -y)*x)
        db=np.mean(2*(y_pred-y))
        
        w -= lr*dw
        b -= lr*db
    return loss_values
```

```python
lrs=[0.0001,0.01,0.1,1.0]
results={}

for lr in lrs:
    results[lr]=train_model(lr)
```

```python
plt.figure(figsize=(10,6))

for lr in lrs:
    plt.plot(results[lr],label=f"LR={lr}")
    
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Learning Rate Experiment: Loss vs Epochs")
plt.legend()
plt.grid(True)
plt.show()
          
```
