```python
import numpy as np
import matplotlib.pyplot as plt

#softmax function
def softmax(x):
    exp_x=np.exp(x-np.max(x))
    return exp_x/np.sum(exp_x)
scores=np.array([1,2,3,4])
y=softmax(scores)

#plot
plt.bar(range(len(scores)),y)
plt.title("Softmax Activation (probability Distribution)")
plt.xlabel("Class")
plt.ylabel("Probability")
plt.grid(True)
plt.show() 

print("\nSoftmax Advantages:")
print("1.Coneverts raw numbers into probabilities.")
print("2.Sum of outputs=1-> interpretable.")

print("\nSoftmax Disadvantages:")
print("1.very sensitive to large inputs (exp grows fast).")
print("2.Not used in hidden layers,only output layer.")
print("3.Can cause numerical overflow.")
```
