```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,400)

#sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
y=sigmoid(x)

#plot
plt.plot(x,y)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

print("\nSigmoid Advantages:")
print("1.output between 0 and 1 -> good for binary classification.")
print("2.Smooth and differentiable everywhere.")

print("\nSigmoid Disadvantages:")
print("1.Causes vanishing gradient when x is very  +ve or -Ve.")
print("2.Not zero-centered -> slows learning.")
print("3.Computation slightly expensive (exp Function).")
```
