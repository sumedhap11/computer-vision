```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,400)

#Relu function
def relu(x):
    return np.maximum(0,x)
y=relu(x)

#plot
plt.plot(x,y)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("output")
plt.grid(True)
plt.show()

print("\nReLU Adavantages:")
print("1.Very Simple and fast to compute.")
print("2.Does not saturate for positive values.")
print("3. Helps deep networkd=s learn faster (no vanishing gradient for x>0.)")

print("\nReLU Disadvantages:")
print("1.for negayive x, output becomes 0-> neuron 'dies.'")
print("2.Not zero-centered.")
print("3.Gradient is 0 for x <0-> learning stops there.")
```
