```python
import numpy as np

x=np.array([[-1.0],
           [-0.5],
           [0.0],
           [0.5],
           [1.0]])
y= 2* x+1
```

```python
#network architecture
#-1 input neuron
#-2 hidden layer neuron
#hidden activation=sigmoid
#loss function = mean Squared Error
np.random.seed(0)
in_dim =1
hidden_dim =2
out_dim = 1
```

```python
W1 = np.random.randn(in_dim, hidden_dim)*0.1
b1 = np.zeros((1,hidden_dim))

W2=np.random.randn(hidden_dim,out_dim)*0.1
b2=np.zeros((1,out_dim))
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
def sigmoid_prime(z):
    s=sigmoid(z)
    return s*(1-s)
```

```python
lr=0.1
epochs=200
```

```python
def mse(y_true,y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
for epoch in range(1,epochs+1):
    z1=x@W1+b1
    a1=sigmoid(z1)
    
    z2=a1@W2+b2
    y_pred=z2
    
    loss=mse(y,y_pred)
    
    N=x.shape[0]
    
    dL_dy=(2.0/N)*(y_pred-y)
    
    dW2=a1.T @ dL_dy
    
    db2=np.sum(dL_dy,axis=0,keepdims=True)
    
    delta_hidden=(dL_dy @ W2.T) * sigmoid_prime(z1)
    
    dW1=x.T @ delta_hidden
    db1=np.sum(delta_hidden,axis=0,keepdims=True)
    
    W2-=lr*dW2
    b2-=lr*db2
    
    if epoch % 20==0 or epoch==1:
        print(f"Epoch{epoch:3d} Loss={loss:.6f}")
        
        print("\nFinal predictions vs true values:")
        
        
z1=x @ W1 +b1
a1=sigmoid(z1)
y_pred = a1 @ W2 + b2

for xi,yi,ypi in zip(x.flatten(),y.flatten(),y_pred.flatten()):
    print(f"x={xi:.2f} true={yi:.2f} pred={ypi:.3f}")
```
