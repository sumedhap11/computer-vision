```python
import numpy as np
#step1: True labels(0 or 1)
y_true=np.array([1,0,1,0])
```

```python
#step2
y_pred=np.array([0.9,0.2,0.8,0.4])
```

```python
#step3: compute BCE manually
bce=-np.mean(y_true *np.log(y_pred)+
            (1-y_true)*np.log(1-y_pred))

print("True lables: ",y_true)
print("Predicted probabilities: ",y_pred)
print("\nBinary Cross Entropy=",bce)
```
